We analyze descriptive statistics and visualize relationships between different factors affecting happiness scores of countries.

## Libraries Used

- NumPy – Numerical operations
- Matplotlib – Basic plots

## Tasks Performed
- Data Loading and Inspection
- Descriptive Statistics
- Histogram
- Box Plot
- Scatter Plot
- Line Plot
- Bar Plot
- Heatmap
- Pair Plot

## Outcome
This notebook helps understand trends, correlations, and distribution of happiness factors across countries.


```python
#Karthick
#24BAD055
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv('happiness_2017.csv')
df.head()
```

Please upload the `happiness_2017.csv` file to your Colab environment using the file icon on the left sidebar. Once uploaded, you can run the following cell to load the data.

```python
df = pd.read_csv('happiness_2017.csv')
df.head()
```

```python
df.info()
df.describe()
```

```python
print("Mean:\n", df.mean(numeric_only=True))
print("\nMedian:\n", df.median(numeric_only=True))
print("\nMode:\n", df.mode(numeric_only=True))
print("\nStandard Deviation:\n", df.std(numeric_only=True))
```

```python
df.hist(figsize=(12,10))
plt.suptitle("Histogram of Numerical Features")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df.select_dtypes(include=np.number))
plt.title("Box Plot")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['HappinessScore'], df['Log GDP per capita'])
plt.xlabel('Happiness Score')
plt.ylabel('GDP per Capita')
plt.title('Happiness vs GDP')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df['HappinessScore'])
plt.title("Line Plot of Happiness Score")
plt.show()
```

```python
top10 = df.head(10)
plt.figure(figsize=(10,6))
plt.bar(top10['Country'], top10['HappinessScore'])
plt.xticks(rotation=45)
plt.title("Top 10 Countries Happiness Score")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df.select_dtypes(include=np.number))
plt.show()
```
