```python
# Assignment 3 - EDA with 30 Different Plots (Car Price Dataset)
# Name: N Karthick
# Roll No: 24BAD055

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

url = "https://raw.githubusercontent.com/selva86/datasets/master/Auto.csv"
df = pd.read_csv(url)

print(df.head())
print(df.info())
print(df.describe())

# Handling missing values
df = df.dropna()

print("\nMissing values after cleaning:\n", df.isnull().sum())
```

Plot 1: Histogram of MPG



This histogram shows the frequency distribution of mileage (MPG) values in the dataset.
It helps identify the most common mileage range and whether the data is skewed or normally distributed.

```python
plt.figure()
plt.hist(df["mpg"], bins=20)
plt.title("1. Histogram of MPG")
plt.xlabel("MPG")
plt.ylabel("Frequency")
plt.show()

print("Description: Histogram shows distribution of MPG and helps identify common mileage range.")
```

Plot 2: KDE Plot of Horsepower

This KDE plot gives a smooth density distribution of horsepower values.
It helps understand where horsepower values are concentrated and how spread out the data is.

```python
plt.figure()
sns.kdeplot(df["horsepower"], fill=True)
plt.title("2. KDE Plot of Horsepower")
plt.xlabel("Horsepower")
plt.show()

print("Description: KDE plot gives smooth density curve and shows how horsepower values are spread.")
```

Plot 3: Boxplot of Weight

This boxplot summarizes the distribution of car weights using median and quartiles.
It is mainly useful to detect outliers and understand the spread of weight values.

```python
plt.figure()
sns.boxplot(x=df["weight"])
plt.title("3. Boxplot of Weight")
plt.show()

print("Description: Boxplot helps detect outliers and shows weight distribution using median and quartiles.")
```

Plot 4: Violin Plot of Acceleration

This violin plot displays the distribution and density of acceleration values.
It helps visualize how acceleration values are spread and where they are more frequent.

```python
plt.figure()
sns.violinplot(x=df["acceleration"])
plt.title("4. Violin Plot of Acceleration")
plt.show()

print("Description: Violin plot shows both distribution shape and spread of acceleration values.")
```

Plot 5: Countplot of Cylinders

This countplot shows how many cars belong to each cylinder category.
It helps identify the most common engine types present in the dataset.

```python
plt.figure()
sns.countplot(x="cylinders", data=df)
plt.title("5. Countplot of Cylinders")
plt.show()

print("Description: Countplot shows how many cars exist for each cylinder category.")
```

Plot 6: Barplot of Average MPG vs Cylinders

This barplot compares the average MPG across different cylinder categories.
It helps analyze how engine size affects fuel efficiency.

```python
plt.figure()
sns.barplot(x="cylinders", y="mpg", data=df)
plt.title("6. Barplot: Average MPG vs Cylinders")
plt.show()

print("Description: Barplot compares average MPG across cylinder categories.")
```

Plot 7: Scatter Plot of Horsepower vs MPG

This scatter plot shows the relationship between horsepower and mileage (MPG).
It helps detect whether MPG decreases as horsepower increases and reveals overall trends.

```python
plt.figure()
plt.scatter(df["horsepower"], df["mpg"])
plt.title("7. Scatter Plot: Horsepower vs MPG")
plt.xlabel("Horsepower")
plt.ylabel("MPG")
plt.show()

print("Description: Scatter plot shows relationship between horsepower and mileage (MPG).")
```

Plot 8: Regression Plot of Horsepower vs MPG

This regression plot includes a best-fit line along with scatter points.
It helps understand the linear trend and how strongly horsepower affects MPG.

```python
plt.figure()
sns.regplot(x="horsepower", y="mpg", data=df)
plt.title("8. Regression Plot: Horsepower vs MPG")
plt.show()

print("Description: Regression plot adds best-fit line to understand trend between horsepower and MPG.")
```

Plot 9: Scatter Plot of Weight vs MPG (Hue = Origin)

This scatter plot shows how weight and MPG are related while differentiating car origins using color.
It helps compare how different regions produce cars with different efficiency patterns.

```python
plt.figure()
sns.scatterplot(x="weight", y="mpg", hue="origin", data=df)
plt.title("9. Scatter Plot: Weight vs MPG (Origin)")
plt.show()

print("Description: Scatter with hue compares mileage across different car origins.")
```

Plot 10: Line Plot of Average MPG Over Years

This line plot shows the trend of average MPG over production years.
It helps observe improvements in fuel efficiency across time due to better technology.

```python
df_year = df.groupby("year")["mpg"].mean()

plt.figure()
plt.plot(df_year.index, df_year.values, marker="o")
plt.title("10. Line Plot: Average MPG Over Years")
plt.xlabel("Year")
plt.ylabel("Average MPG")
plt.show()

print("Description: Line plot shows how average MPG changed over time.")
```

Plot 11: Area Plot of MPG Trend Over Years

This area plot highlights the MPG trend over years with a filled region.
It is useful for emphasizing growth or decline patterns in average mileage over time.

```python
plt.figure()
plt.fill_between(df_year.index, df_year.values)
plt.title("11. Area Plot: MPG Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Average MPG")
plt.show()

print("Description: Area plot highlights overall increase or decrease of MPG over years.")
```

Plot 12: Pie Chart of Origin Distribution

This pie chart represents the percentage of cars belonging to each origin category.
It helps understand dataset composition and which origin dominates the dataset.

```python
plt.figure()
df["origin"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("12. Pie Chart: Car Origin Distribution")
plt.ylabel("")
plt.show()

print("Description: Pie chart shows percentage distribution of cars by origin.")
```

Plot 13: Strip Plot of MPG vs Cylinders

This strip plot displays individual MPG values for each cylinder category.
It helps observe variation in MPG and highlights overlapping values in each class.

```python
plt.figure()
sns.stripplot(x="cylinders", y="mpg", data=df, jitter=True)
plt.title("13. Strip Plot: MPG vs Cylinders")
plt.show()

print("Description: Strip plot shows individual MPG points for each cylinder category.")
```

Plot 14: Swarm Plot of MPG vs Cylinders

This swarm plot shows MPG distribution per cylinder category without overlapping points.
It provides a clearer view of data clustering compared to strip plots.

```python
plt.figure()
sns.swarmplot(x="cylinders", y="mpg", data=df)
plt.title("14. Swarm Plot: MPG vs Cylinders")
plt.show()

print("Description: Swarm plot avoids overlap and clearly shows distribution of MPG per cylinders.")
```

Plot 15: Boxplot of MPG vs Cylinders

This boxplot compares MPG distribution across cylinder categories.
It helps identify which cylinder group has higher median MPG and which group contains outliers.

```python
plt.figure()
sns.boxplot(x="cylinders", y="mpg", data=df)
plt.title("15. Boxplot: MPG vs Cylinders")
plt.show()

print("Description: Boxplot compares MPG spread and outliers for each cylinder group.")
```

Plot 16: Violin Plot of MPG vs Origin

This violin plot shows MPG distribution for each origin group.
It helps compare the density and spread of MPG values among different origins.

```python
plt.figure()
sns.violinplot(x="origin", y="mpg", data=df)
plt.title("16. Violin Plot: MPG vs Origin")
plt.show()

print("Description: Violin plot compares MPG distribution for each origin category.")
```

Plot 17: Pairplot of Important Numerical Features

This pairplot shows scatter plots between multiple numerical features together.
It helps identify relationships, trends, and possible correlations across many variables at once.

```python
sns.pairplot(df[["mpg", "horsepower", "weight", "acceleration", "cylinders"]])
plt.suptitle("17. Pairplot of Important Features", y=1.02)
plt.show()

print("Description: Pairplot shows multiple scatter plots and helps understand feature relationships.")
```

Plot 18: Correlation Heatmap

This heatmap displays correlation values between numerical features using color intensity.
It helps identify which features are strongly related and can affect regression models.

```python
plt.figure(figsize=(10,6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("18. Correlation Heatmap")
plt.show()

print("Description: Heatmap shows correlation strength between numerical variables.")
```

Plot 19: Missing Values Heatmap

This heatmap visually represents missing values across the dataset.
It helps quickly identify which columns contain null values and the pattern of missingness.

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False)
plt.title("19. Missing Values Heatmap")
plt.show()

print("Description: Missing value heatmap shows if any null values exist in the dataset.")
```

Plot 20: Joint Plot of Weight vs MPG

This joint plot combines a scatter plot with histograms on both axes.
It helps understand both the relationship between variables and their individual distributions.

```python
sns.jointplot(x="weight", y="mpg", data=df, kind="scatter")
plt.suptitle("20. Joint Plot: Weight vs MPG", y=1.02)
plt.show()

print("Description: Joint plot shows scatter plot with histograms on both axes for better distribution analysis.")
```

Plot 21: Hexbin Plot of Weight vs MPG

This hexbin plot groups dense scatter points into hexagonal bins.
It is useful when data points are too many and helps show density concentration clearly.

```python
plt.figure()
plt.hexbin(df["weight"], df["mpg"], gridsize=30)
plt.colorbar(label="Count")
plt.title("21. Hexbin Plot: Weight vs MPG")
plt.xlabel("Weight")
plt.ylabel("MPG")
plt.show()

print("Description: Hexbin plot is useful when there are many points and helps show density clusters.")
```

Plot 22: ECDF Plot of MPG

This ECDF plot shows the cumulative distribution of MPG values.
It helps understand percentile-based analysis such as median MPG and probability distribution.

```python
plt.figure()
sns.ecdfplot(df["mpg"])
plt.title("22. ECDF Plot of MPG")
plt.xlabel("MPG")
plt.ylabel("ECDF")
plt.show()

print("Description: ECDF plot shows cumulative distribution and helps compare percentile-based values.")
```

Plot 23: Rug Plot of Horsepower

This rug plot shows individual horsepower values as small ticks along the axis.
It helps visualize density and clustering of horsepower values in the dataset.

```python
plt.figure()
sns.rugplot(df["horsepower"])
plt.title("23. Rug Plot of Horsepower")
plt.xlabel("Horsepower")
plt.show()

print("Description: Rug plot shows small tick marks for each data point and helps visualize data density.")
```

Plot 24: QQ Plot of MPG

This QQ plot compares the MPG distribution with a normal distribution.
It helps check whether MPG values follow a normal pattern or show deviation.

```python
import scipy.stats as stats

plt.figure()
stats.probplot(df["mpg"], dist="norm", plot=plt)
plt.title("24. QQ Plot of MPG")
plt.show()

print("Description: QQ plot checks whether MPG values follow a normal distribution.")
```

Plot 25: Stacked Histogram of MPG by Origin

This stacked histogram shows MPG distribution grouped by origin categories.
It helps compare how mileage values differ between different regions.

```python
plt.figure()
sns.histplot(data=df, x="mpg", hue="origin", multiple="stack")
plt.title("25. Stacked Histogram: MPG by Origin")
plt.show()

print("Description: Stacked histogram compares MPG distribution across origins.")
```

Plot 26: KDE Plot of MPG by Origin

This KDE plot shows smooth MPG distributions for each origin group.
It helps understand which origin has better mileage and how distributions overlap.

```python
plt.figure()
sns.kdeplot(data=df, x="mpg", hue="origin", fill=True)
plt.title("26. KDE Plot: MPG by Origin")
plt.show()

print("Description: KDE plot by origin shows mileage distribution differences across countries.")
```

Plot 27: Clustermap of Correlation

This clustermap groups correlated features into clusters automatically.
It helps identify groups of features that behave similarly and may cause multicollinearity.

```python
sns.clustermap(corr, annot=True, cmap="coolwarm")
plt.title("27. Cluster Map of Correlation")
plt.show()

print("Description: Clustermap groups similar features together and shows correlation clusters.")
```

Plot 28: Boxen Plot of Horsepower vs Origin

This boxen plot is an advanced version of boxplot for detailed distribution.
It helps analyze horsepower spread and variation among different origins.

Plot 29: Catplot of MPG vs Cylinders (Grouped by Origin)

This categorical plot compares average MPG across cylinders and origin categories.
It helps understand how both engine size and origin influence mileage together.

```python
sns.catplot(x="cylinders", y="mpg", hue="origin", data=df, kind="bar")
plt.title("29. Catplot: MPG vs Cylinders (Origin)")
plt.show()

print("Description: Catplot compares categorical and numerical features with grouping.")
```

Plot 30: 3D Scatter Plot (Horsepower vs Weight vs MPG)

This 3D scatter plot shows the relationship among three numerical features at once.
It helps visualize how horsepower and weight together impact the mileage (MPG).

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

ax.scatter(df["horsepower"], df["weight"], df["mpg"])
ax.set_xlabel("Horsepower")
ax.set_ylabel("Weight")
ax.set_zlabel("MPG")

plt.title("30. 3D Scatter Plot: Horsepower vs Weight vs MPG")
plt.show()

print("Description: 3D scatter plot helps visualize relationship among three numerical features simultaneously.")
```
